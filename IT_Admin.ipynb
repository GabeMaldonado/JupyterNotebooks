# Different topics on IT Admin (Windows)

### Checking All Windows Services on a Remote Machine

*   Open CMD as an admin
*   sc \\IP_or_machine_name query state= all
    *   ```C:\sc \\10.10.10.10 query state= all```

This will return all the services, running, stopped, disabled.
The list contains all the info about the service such as the SERVICE_NAME, DISPLAY_NAME, TYPE, & STATE.

````
from IPython.display import Image
Image(filename='/content/List_Services.PNG')
````

### Starting/Stopping a SERVICE on a Remote Client

We can find the desired service from the output list of the comamand above and we can start/stop it.

*   ```C:\>sc \\10.10.10.10 start "SERVICE_NAME"```

We can aslo change the SERVICE's start type to auto, auto delayed and disabled.

*   ```C:\>sc \\10.10.10.10 config "SERVICE_NAME" start= auto```


````
from IPython.display import Image
Image(filename='/content/Start_Service.PNG')
````

### Connecting to a Network Registry

If the 'RemoteRegistry' is enabled and running on the remote machine, we can connect to its registry from your local machine.

*   Open the registry:
    
    *   WIN + R --> regedit.exe

*   File\Connect Network Registry
*   Enter the IP or machine name
*   Enter credentials

If the connection was successfull we will see the remote host IP at the bottom of the list. The HKEY_LOCAL_MACHINE and HKEY_USERS we can browse and make changes to the registry.

````
from IPython.display import Image
Image(filename='/content/Remote_Registry.PNG')
````

### Killing a proceess remotely

We can use ```taskkill``` to kill a task/process on a remote machine. For this we need the remote machine name/IP username and password as well as knowing the proceess we wish to terminate:

*   ```C:\>taskkill /S 10.10.10.10 /U username /P password /IM processname```
