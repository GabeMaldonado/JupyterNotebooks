# Configuring a Python Virtual Environment in VSCode

When working in a project it is recommended to work inside a virtual environment rather than installing packages locally. If we are collaborating with other people it is better to have all the packages contained within the project rather than have the files locally in every single machine so that way everyone working on the project will have the same requirements. 


*   Create a folder for your project, i.e. FlaskIntro
*   Open folder on VS code
*   CRTL + ~ to open the shell
*   Optional -- if you don't have virtualenv in the machine install it by running ```pip3 install virtualenv``` on the shell
*   Create virtual environment by running ```virtual env``` -- *env* is a the default name
*   Once the virtual env is created we need to activate it by running:
    *   On Mac - ```$ source env/bin/activate```
    *   On Windows - ```env\Scripts\activate.bat``` 
*   Once we see the ```(env)``` it means that we have successfully created and activated the virtual env and that we are now working inside the virtual environment.
